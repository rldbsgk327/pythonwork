```python
##### 전자 계산기 프로그램(사칙연산)


"""
----------------------------------
연산자 : +
숫자1 : 10
숫자2 : 5

결과 : 10 + 5 = 15


전자 계산기 프로그램(사칙연산)   -> chat올리기
----------------------------------
연산자 : +
숫자1 : 10
숫자2 : 5

결과 : 10 + 5 = 15

전자 계산기 프로그램(사칙연산)   -> chat올리기
----------------------------------
연산자 : +
숫자1 : 10
숫자2 : 5

결과 : 10 + 5 = 15

"""



op = input("연산자를 입력하세요: ")
num1 = int(input("숫자를 입력하세요: "))
num2 = int(input("숫자를 입력하세요: "))

if op == '+':
    print(num1 + num2)
elif op == '-':
    print(num1 - num2)
elif op == '*':
    print(num1 * num2)
elif op == '/':
    print(num1 / num2)
else: print("연산자가 잘못되었습니다. +, -, *, / 중에서 입력해주세요")
```

```python
##### 문제 1: 전자계산기 업그레이드

"""

전자계산기 프로그램을 반복문을 이용하여 업그레이드
(끝나는 조건은 연산자에 x를 입력하면 종료)

..전자계산기 문제에 접목하기...true,종료문

"""

while True:

    op = input("연산자를 입력하세요: ")
    
    while op != 'x':
        num1 = int(input("숫자를 입력하세요: "))
        num2 = int(input("숫자를 입력하세요: "))
        break

    if op == '+':
        print(num1 + num2)
    elif op == '-':
        print(num1 - num2)
    elif op == '*':
        print(num1 * num2)
    elif op == '/':
        print(num1 / num2)
    elif op == 'x':
        break
    else: print("연산자가 잘못되었습니다. +, -, *, / 중에서 입력해주세요")

```

```python
###### 문제 3: 짝수끼리, 홀수끼리의 합

"""

숫자를 입력받아 짝수의 합과 홀수의 합을 구하시오.(0을 입력하면 종료)

숫자1: 2
숫자2: 3
숫자3: 1
숫자4: 6
숫자5: 0

짝수의 합:?
홀수의 합:?

"""

cout = 1
odd = 0
even = 0

while True:
    
    num = int(input("숫자{}:".format(cout) ))
    cout += 1
    
    if num % 2 == 0:
        odd += num
        if num == 0:
            break
        
    else:
        even += num
        
        
print()
print("짝수의 합:", even)
print("홀수의 합:", odd)


```

```python
##### 문제 5: 친구 관리 프로그램

"""
친구 관리 프로그램  /////1-9메뉴는 계속 반복해서 출력

====================

1.친구 리스트 출력
2.친구 추가 ...list 사용
3.친구 삭제
4.이름 변경
9.종료
메뉴를 선택 : 2 (엔터)
이름을 입력 : 홍길동 (엔터)

-----------------------
1.친구 리스트 출력
2.친구 추가
3.친구 삭제
4.이름 변경
9.종료
메뉴를 선택 : 2 (엔터)
이름을 입력 : 임꺽정 (엔터)

-----------------------
1.친구 리스트 출력
2.친구 추가
3.친구 삭제
4.이름 변경
9.종료
메뉴를 선택 : 1 (엔터)


["홍길동", "임꺽정"]

----------------------
1.친구 리스트 출력
2.친구 추가
3.친구 삭제
4.이름 변경
9.종료
메뉴를 선택 : 4 (엔터)
변경할 이름 선택 : 홍길동 (엔터)
변경하고 싶은 이름 입력 : 홍길순 (엔터)

-------------------------------
1.친구 리스트 출력
2.친구 추가
3.친구 삭제
4.이름 변경
9.종료
메뉴를 선택 : 1 (엔터)
["홍길순", "임꺽정"]

-------------------------------
1.친구 리스트 출력
2.친구 추가
3.친구 삭제
4.이름 변경
9.종료
메뉴를 선택 : 3 (엔터)
삭제할 이름 입력 : 홍길순

------------------------------
"""


a = list()

print("친구관리 프로그램")
print("=================================")

while True :
    print(" 1.친구 리스트 출력\n 2.친구 추가\n 3.친구 삭제\n 4.이름 변경\n 9.종료")
    menu = int(input(" 메뉴를 선택: "))
    
    if menu == 1:
        print(a)
        print("-------------------------------------")
        
    if menu == 2:
        name = input(" 이름을 입력:")
        a.append(name)
        print("-------------------------------------")
        
    if menu == 3:
        name = input(" 이름을 입력:")
        a.remove(name)
        print("-------------------------------------")
    
    if menu == 4:
        name_re = input(" 변경할 이름 입력:")
        name_ne = input(" 변경하고 싶은 새로운 이름 입력:")
        b = a.index(name_re)
        a[b] = name_ne
        print("-------------------------------------")
    
    if menu == 9:
        print("-------------------------------------")
        break
```
